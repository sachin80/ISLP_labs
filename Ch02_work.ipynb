$\text{Inner\hspace{0.1cm}product:}$
$$
x^Ty=
\begin{pmatrix}
x_1 & x_2 & \dots & x_n
\end{pmatrix}
\begin{pmatrix}
y_1\\
y_2\\
\vdots\\
y_n
\end{pmatrix}
 = x_1 y_1 + x_2y_2 + \dots x_ny_n
$$

$\text{Outer\hspace{0.1cm}product:}$

$$
xy^T=
\begin{pmatrix}
x_1\\
x_2\\
\vdots\\
x_n
\end{pmatrix}
\begin{pmatrix}
y_1 & y_2 & \dots & y_n
\end{pmatrix} =
\begin{pmatrix}
x_1y_1 & x_1y_2 & \dots & x_1y_n\\
x_2y_1 & x_2y_2 & \dots & x_2y_n\\
\vdots & \vdots & \ddots & \vdots \\
x_ny_1 & x_ny_2 & \dots & x_ny_n\\
\end{pmatrix} = 
xy_1 + xy_2 + \dots + xy_n
$$

```python
# let's do a contour plot
fig, ax = subplots(figsize=(8, 8))
x = np.linspace(-np.pi, np.pi, 50)
y = x
# need to compute hadamard/outer product/for every pair 
f = np.multiply.outer(np.cos(y), 1 / (1 + x**2))
ax.contour(x, y, f)
```

```python
# increase resolution by adding more levels to the countour
ax.contour(x, y, f, levels=50)
fig 
```

```python
# indexing arrays 
# let's create a 4 by 4 array
A = np.array(np.arange(16)).reshape((4, 4))
A 
```

```python
A[1, 2]
```

```python
# to select multiple rows 
A[[1, 3]]
```

```python
# to select first and third cols
A[:, [0, 2]]
```

```python
# we want the submatrix made up of the second and forth rows
A[[1, 3], [0, 2]]  # this is incorrect
```

numpy interprets the two lists as pairwise indices, why they must have the same length. A solution is create a submatrix by subsetting rows of A, then make a further submatrix by subsetting the columns.

```python
A[[1, 3]][:, [0, 2]]
```

```python
# can use the convenience function that creates a intermediate mesh object
idx = np.ix_([1, 3], [0, 2, 3])
A[idx]
```

```python
# can also use slices
A[1:4:2, 0:3:2]
```

Why are we able to retrieve a submatrix directly using slices but not using lists?
Its because they are different `Python` types, and
are treated differently by `numpy`.
Slices can be used to extract objects from arbitrary sequences, such as strings, lists, and tuples, while the use of lists for indexing is more limited.

```python
print(A)
print(A[0])
print(A[[0, 1]])
print(A[0][[0, 1]])
print(A[[0, 1]][:, [0, 1]])
```

```python
# Boolean indexing 
print(A.shape[0])
keep_rows  = np.zeros(A.shape[0], bool)
keep_rows 
```

```python
keep_rows[[1, 3]] = True 
keep_rows
```

```python
# True = 1 and False = 0 
np.all(keep_rows == np.array([0, 1, 0, 1]))
```

(Here, the function `np.all()` has checked whether
all entries of an array are `True`. A similar function, `np.any()`, can be used to check whether any entries of an array are `True`.)

However, even though `np.array([0,1,0,1])`  and `keep_rows` are equal according to `==`, they index different sets of rows!
The former retrieves the first, second, first, and second rows of `A`. 

```python
print(A)
# This will give us the 1st, 2nd, 1st, 2nd rows 
A[np.array([0, 1, 0, 1])]
```

```python
print(A)
print(keep_rows)
# This will give us the 2nd and 4th rows
A[keep_rows]
```

This shows that Booleans and integers are treated differently for indexing. 

```python
# using np.ix_()
keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0, 2, 3]] = True 
idx_bool = np.ix_(keep_rows, keep_cols)
A[idx_bool]
```

```python
# can also mix list with array of Bools 
idx_mixed = np.ix_([1, 3], keep_cols)
A[idx_mixed]
```

1a. A more flexible model is good because $n$ is large then we have many observations. A more flexible model will have lower bias but higher variance and because you have many observations and few predictors you can afford higher variance since you have enough data to reliably estimate the model parameters and less risk of overfitting. 

1b. A more flexible model is bad because $n$ is small then we have less observations. A more flexible model will risk overfitting the data since there might not be enough data points to reliably estimate all the parameters. Usually, it's better to use a less flexible model and most will resort to regularization or simplifying assumptions. 

1c. A more flexible model is good since an inflexible model like a linear regression will not be able to fit a non-linear relationship. 

1d. A more flexible model is probably bad since the variance is already high.
